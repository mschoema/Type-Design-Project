```python
import re
import csv
import requests
from bs4 import BeautifulSoup
from urllib import parse

class Radical(object):
    dictionary_filepath = 'xinhua.csv'
    local = True

    def __init__(self):
        self.read_dictionary()

        self.origin_len = len(self.dictionary)

    def read_dictionary(self):
        self.dictionary = {}

        file = open(self.dictionary_filepath, 'rU')
        reader = csv.reader(file)

        for line in reader:
            word = line[0].encode('utf-8').decode('utf-8')
            radical = line[1].encode('utf-8').decode('utf-8')
            stroke_count = line[2].encode('utf-8').decode('utf-8')
            
            self.dictionary[word] = {'radical':radical,'stroke_count':stroke_count}

        file.close()

    def write_dictionary(self):
        file_obj = open(self.dictionary_filepath, 'w')

        writer = csv.writer(file_obj)
        for word in self.dictionary:
            writer.writerow([word,self.dictionary[word]['radical'],self.dictionary[word]['stroke_count']])

        file_obj.close()

    def get_radical(self,word):
        word = word.encode('utf-8').decode('utf-8')

        if word in self.dictionary and self.local:
            return self.dictionary[word]
        else:
            ret = None
            
            try:
                ret = self.get_radical_from_hanzizidian(word)
            except Exception as e:   
                ret = self.get_radical_from_baiduhanyu(word)
            
            self.add_in_dictionary(word,ret['radical'],ret['stroke_count'])
            return ret

    def post_baidu(self,url):
        print(url)
        try:
            headers = {
                'user-agent': 'Mozilla/5.0 (compatible; MSIE 9.0; Windows NT 6.1; Trident/5.0;',
                'connection': 'keep-alive',
                'referer':url
            }
            r = requests.get(url,headers=headers,timeout=4)
            html = r.text
            html = html.encode('ISO-8859-1').decode('utf8')
            return html
        except Exception as e:
            print( 'URL Request Error:', e)
            return None

    def add_in_dictionary(self,word,radical,stroke_count):
        # add in file
        file_object = open(self.dictionary_filepath,'a+')
        file_object.write(word+','+radical+','+stroke_count+'\r\n')
        file_object.close()


    def get_radical_from_baiduhanyu(self,word):
        url = 'http://hanyu.baidu.com/zici/s?ptype=zici&wd=%s' % parse.quote(word)
        html = self.post_baidu(url)

        if html == None:
            return None
        
        def anlysis_radical_from_html(html_doc):
            soup = BeautifulSoup(html_doc, 'html.parser')

            radical = soup.find(id="radical").span.contents[0]
            stroke_count = soup.find(id="stroke_count").span.contents[0]

            return {'radical':radical,'stroke_count':stroke_count}

        ret = anlysis_radical_from_html(html)
        if ret['radical'] != None and ret['stroke_count'] != None:
            self.dictionary[word] = ret

        return ret

    def post_hanzizidian(self,url):
        print(url)
        try:
            headers = {
                'user-agent': 'Mozilla/5.0 (compatible; MSIE 9.0; Windows NT 6.1; Trident/5.0;',
                'connection': 'keep-alive',
                'referer':url
            }
            r = requests.get(url,headers=headers,timeout=4)
            html = r.text
            return html
        except Exception as e:
            print( 'URL Request Error:', e)
            return None

    def get_radical_from_hanzizidian(self,word):
        url = 'http://www.hanzizidian.com/lookup.php?w=%s' % parse.quote(word)
        html = self.post_hanzizidian(url)

        if html == None:
            return None

        def anlysis_radical_from_html(html_doc):
            soup = BeautifulSoup(html_doc, 'html.parser')

            txt = soup.find_all(attrs={"class":"dataft"})[-1].parent.text
            print(txt)

            matchObj = re.search(r'(简体部首|^部首):([\w]+),部外:([\d]+),总:([\d]+)', txt)

            radical = matchObj.group(2)
            stroke_count = matchObj.group(4)

            return {'radical':radical,'stroke_count':stroke_count}

        ret = anlysis_radical_from_html(html)
        if ret['radical'] != None and ret['stroke_count'] != None:
            self.dictionary[word] = ret

        return ret


    def save(self):
        if len(self.dictionary) > self.origin_len:
            self.write_dictionary()
```

```python
import functools

def takeStrokeCount(elem):
    return int(radical.dictionary[elem]['stroke_count'])

def radicalDict():
    def takeMx(e1,e2):
        if takeStrokeCount(e1) == takeStrokeCount(e2):
            return takeUnicode(e1) < takeUnicode(e2) if -1 else 1
        else:
            return takeStrokeCount(e1) - takeStrokeCount(e2)

    bssorted.sort(key=functools.cmp_to_key(takeMx))
    dicts = []
    maps = {}
    index = 0
    
    for bs in bssorted:
        no = "rid"+str(index).zfill(3)
        dicts.append([
            no,
            bs,
            takeStrokeCount(bs),
            takeUnicode(bs)[2:]
        ])
        maps[bs] = no
        index+=1
        
    return (dicts,maps)
    
radical_dict = radicalDict()
radical_dict[1]
```
