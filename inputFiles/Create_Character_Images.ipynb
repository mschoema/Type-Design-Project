```python
import matplotlib.pyplot as plt

import os
from PIL import Image, ImageFont, ImageDraw, ImageChops

# 图片大小，像素
size = 1000
# 字体文件名称
fonts = ["HYQiHei-25JF.otf", "HanYiXiXingKaiJian-1.ttf", "Kaiti_2.ttf", "SourceHanSansCN-Regular.otf", "HYXinRenWenSongW-1.otf","SourceHanSerifCN-Heavy-4.otf"]
styles = ["Heiti", "Kaiti", "Kaiti", "Regular", "Songti", "Songti"]
adds = ["_2", "_1", "_2", "", "_1", "_2"]
offsets = [-70, 0, -20, 0, -60, -260]
assert(len(fonts)==len(styles))
assert(len(styles)==len(adds))
assert(len(adds)==len(offsets))

def getFontImage(text,size,style,font,y_offset):
    im=Image.new("RGB",(size,size),(255,255,255))
    dr=ImageDraw.Draw(im)

    font=ImageFont.truetype(os.path.join("./",style,font),size)
    dr.text((0,y_offset),text,font=font,fill="#000000")
    im = im.convert('1')
    return im

def isEmpty(im):
    return not ImageChops.invert(im).getbbox()
```

```python
font = fonts[5]
style = styles[5]
y_offset = offsets[5]
character = '丢'

# 单独一个字保存到gb2312目录下
unicode = takeUnicode(character)[2:]
print(unicode)
im = getFontImage(character,size,style,font,y_offset)

print(isEmpty(im))

# 单独预览一个字
plt.imshow(im)
plt.axis('on') # 关掉坐标轴为 off
plt.show()
```

```python
for i in range(len(fonts)):
    style = styles[i]
    add = adds[i]
    write_directory = os.path.dirname('../characterImages2/{}/'.format(style+add))
    if not os.path.exists(write_directory):
        os.makedirs(write_directory)
```

```python
for i in range(len(fonts)):
    font = fonts[i]
    style = styles[i]
    add = adds[i]
    y_offset = offsets[i]
    for char in char_sorted:
        unicode = takeUnicode(char)[2:] 
        try:
            im = getFontImage(char,size,style,font,y_offset)
            if isEmpty(im):
                print("Character missing for font: " + style + "/" + font)
                print("Unicode: ", unicode)
            else:
                im.save('../characterImages2/{}/{}.png'.format(style+add,unicode))
        except Exception as error:
            print("Exception: ", error)
            print("For unicode: ", unicode)
```
